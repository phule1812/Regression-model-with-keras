# Build a Regression Model in Keras

```python
import pandas as pd
import numpy as np
```

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
concrete_data.shape
```

```python
concrete_data.describe()
```

```python
concrete_data.isnull().sum()
```

```python
#Split data into predictors and target
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] 
```

```python
predictors.head()
```

```python
target.head()
```

```python
n_cols = predictors.shape[1]
```

# A. Build a baseline model

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
#define the model
def regression_model():
    model = Sequential()
    model.add(Dense(10, activation = 'relu', input_shape=(n_cols,)))
    model.add(Dense(1))

    model.compile(optimizer= 'adam', loss = 'mean_squared_error')
    return model
```

```python
#spliting data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(predictors, target, test_size = 0.3)
```

```python
model = regression_model()
```

```python
model.fit(x_train, y_train, epochs = 50)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
yhat = model.predict(x_test)
error = mean_squared_error(y_test, yhat)
error
```

```python
list_error = []
for i in range(50):
    model.fit(x_train, y_train, epochs = 50)
    yhat = model.predict(x_test)
    error = mean_squared_error(y_test, yhat)
    list_error.append(error)
    
```

```python
print('Mean of the mean squared errors : ', np.mean(list_error))
print('Standard deviation of the mean squared errors : ', np.std(list_error))
```

# B. Normalize the data

```python
#function for training and creation list of errors
def list_of_error(predictors, target, epochs):
    x_train, x_test, y_train, y_test = train_test_split(predictors, target, test_size = 0.3)
    list_error = []
    for i in range(50):
        model.fit(x_train, y_train, epochs = epochs)
        yhat = model.predict(x_test)
        error = mean_squared_error(y_test, yhat)
        list_error.append(error)
    return list_error
```

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
list_error_norm = list_of_error(predictors_norm, target, 50)
```

```python
print('Mean of the mean squared errors with normalization: ', np.mean(list_error_norm))
print('Standard deviation of the mean squared errors with normalization: ', np.std(list_error_norm))
```

Clearly, the mean of mean squaed errors are smaller compared to that from step A. From that, we can know that with normalization, the neural network will work better.

# C. Increate the number of epochs

```python
list_error_c = list_of_error(predictors_norm, target, 100)
```

```python
print('Mean of the mean squared errors with increation the number of epochs: ', np.mean(list_error_c))
print('Standard deviation of the mean squared errors with increation the number of epochs: ', np.std(list_error_c))
```

From this can see that, the means of mean squared errors from step C and step B are not too different. So, with eppochs 50 from step B is good enough for our model.
